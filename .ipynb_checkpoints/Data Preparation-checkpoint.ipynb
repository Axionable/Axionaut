# Axionaut V2 - Data Preparation
This notebook show you how to create a balanced dataset by selecting the desired amount of examples per class using Pandas.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

## Defining Functions

```python

def label_decoder(line):
    '''Returns human-readible label from target array.
    
    '''
    if np.array_equal(line, [1,0,0,0,0]):
        return 'hard_right'
    elif np.array_equal(line, [0,1,0,0,0]):
        return 'soft_right'
    elif np.array_equal(line, [0,0,1,0,0]):
        return 'straight'
    elif np.array_equal(line, [0,0,0,1,0]):
        return 'soft_left'
    elif np.array_equal(line, [0,0,0,0,1]):
        return 'hard_left'
    else:
        print('Error: Label not recognized.')
        return 

    
def label_decoder_int(line):
    '''Convert label from array to integer.
    
    '''
    if np.array_equal(line, [1,0,0,0,0]):
        return -1
    elif np.array_equal(line, [0,1,0,0,0]):
        return -0.5
    elif np.array_equal(line, [0,0,1,0,0]):
        return 0
    elif np.array_equal(line, [0,0,0,1,0]):
        return 0.5
    elif np.array_equal(line, [0,0,0,0,1]):
        return 1
    else:
        print('Error: Label not recognized.')
        return 
```

## Loading Data

```python
# Load dada from Numpy array
X = np.load('ironcar_labelizada/X_pista.npy')
Y = np.load('ironcar_labelizada/Y_pista.npy')
print('Ironcar data Loaded. Shape = ', np.shape(X))
print('Ironcar data Loaded. Shape = ', np.shape(Y))
```

## Data Visualisation

```python
plt.imshow(X[145])
plt.title(label_decoder(Y[145]))
```

```python
plt.imshow(X[14000])
plt.title(label_decoder(Y[14000]))
```

## Labels Exploration
Using Pandas, we can easily know the global distribution of the data. 


```python
# Convert targets to Numpy arrays
df = pd.DataFrame(Y)

# Convert targets to readible targets (-1 = hard_right, 1 = hard_left)
df['direction'] = df.apply(label_decoder_int, axis=1)

df.describe()
```

```python
df.direction.plot.hist()
plt.title('Non-balanced Histogram')
```

### Note:
Data distribution is not well balanced. It is necessary to equilibrate the number of examples per class.

## Balancing the Dataset
Using Pandas, we can create small data frames containing only a random selection of each group as desired.

```python
# Take a sub-set of soft_left examples (3000 randomly selected)
soft_left = df[df.direction == 0.5].sample(n=3000)
# Take a sub-set of hard_left examples (1000 randomly selected)
hard_left = df[df.direction == 1].sample(n=1000)

# Repeat the same with other classes.
soft_right = df[df.direction == -0.5].sample(n=3000)
hard_right = df[df.direction == -1].sample(n=1000)
straight = df[df.direction == 0]

# Extracting the desired images using its indexes
soft_left_X = X[soft_left.index]
soft_left_Y = Y[soft_left.index]

hard_left_X = X[hard_left.index]
hard_left_Y = Y[hard_left.index]

soft_right_X = X[soft_right.index]
soft_right_Y = Y[soft_right.index]

straight_X = X[straight.index]
straight_Y = Y[straight.index]

hard_right_X = X[hard_right.index]
hard_right_Y = Y[hard_right.index]

# Concatenate the balanced dataset
balanced_X = np.concatenate((soft_left_X, hard_left_X, 
                             soft_right_X, straight_X,
                             hard_right_X))

balanced_Y = np.concatenate((soft_left_Y, hard_left_Y, 
                             soft_right_Y, straight_Y,
                             hard_right_Y))


np.save('balanced_X.npy', balanced_X)
np.save('balanced_Y.npy', balanced_Y)
```

```python
# Convert targets to Numpy arrays
balanced_df = pd.DataFrame(balanced_Y)
#straight_df = pd.DataFrame(Y_straight)

# Convert targets to readible targets (-1 = hard_right, 1 = hard_left)
balanced_df['direction'] = balanced_df.apply(label_decoder_int, axis=1)

balanced_df.direction.plot.hist()
plt.title('Balanced Histogram')
```

### Note:
The final data have a more balanced proportion of examples, according the behaviour that we want the vehicle to have.
